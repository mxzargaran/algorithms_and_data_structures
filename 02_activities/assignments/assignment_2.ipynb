```python
#### v1
# Definition for a binary tree node.
class TreeNode(object):
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

from collections import deque

def is_duplicate(root: TreeNode) -> int:
    if not root:
        return -1

    # Use a set to keep track of visited values
    visited = set()
    
    # Queue for BFS traversal
    queue = deque([root])

    while queue:
        node = queue.popleft()
        if node.val in visited:
            return node.val
        visited.add(node.val)
        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

    return -1

# Helper function to build a tree from a list
def build_tree(values):
    if not values:
        return None

    nodes = []
    for val in values:
        if val is not None:
            nodes.append(TreeNode(val))
        else:
            nodes.append(None)
    
    # Reverse the list
    kids = []
    for i in range(len(nodes) - 1, -1, -1):
        kids.append(nodes[i])
    
    # Set the root as the last element in reversed kids list
    root = kids.pop()

    for node in nodes:
        if node:
            if kids:
                node.left = kids.pop()
            if kids:
                node.right = kids.pop()

    return root

# Test the function 
tree1 = build_tree([1, 2, 2, 3, 5, 6, 7]) # This is Breadth First Search (BFS) traversal method
tree2 = build_tree([1, 10, 2, 3, 10, 12, 12])
tree3 = build_tree([10, 9, 8, 7])

print(f"Example 1: ", is_duplicate(tree1))
print(f"Example 2: ", is_duplicate(tree2))
print(f"Example 3: ", is_duplicate(tree3))
```


-   Explain why their solution works in your own words.


```python
# BFS processes nodes level by level, guaranteeing we find the duplicate closest to the root first
# A visited set tracks seen values, making duplicate detection immediate
# The function returns as soon as a duplicate is found, ensuring the closest one is returned
# If no duplicates exist or the tree is empty, it correctly returns -1
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Time : O(n) => BFS visits every node in the tree where n is the number of nodes
# Space: O(n) => visited set stores at most n values in the worst case
```
